```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data_num = 1000
rssi = pd.read_csv('./normal.csv', header=None)
rssi = rssi.rename({0: 'id_normal', 1: 'rssi_normal'}, axis=1)
# print(rssi)
print("mean rssi: "+str(np.mean(rssi['rssi_normal'])))
print("packet loss probbility: "+str((1-rssi.count()['id_normal']/data_num)*100)+"%")
plt.figure()
rssi['rssi_normal'].plot()
plt.show()
rssi['rssi_normal'].plot.box()
plt.show()
```
