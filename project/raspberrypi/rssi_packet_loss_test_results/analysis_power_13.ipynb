```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data_num = 100
rssi_0 = pd.read_csv('./-0dBm_power_13.csv', header=None)
rssi_0 = rssi_0.rename({0: 'id_0', 1: 'rssi_0'}, axis=1)
# print(rssi_0)
print("mean rssi: "+str(np.mean(rssi_0['rssi_0'])))
print("packet loss probbility: "+str((1-rssi_0.count()['id_0']/data_num)*100)+"%")
plt.figure()
rssi_0['rssi_0'].plot()
plt.show()
rssi_0['rssi_0'].plot.box()
plt.show()
```

```python
data_num = 100
rssi_90 = pd.read_csv('./-90dBm_power_13.csv', header=None)
rssi_90 = rssi_90.rename({0: 'id_90', 1: 'rssi_90'}, axis=1)
print(rssi_90)
print("mean rssi: "+str(np.mean(rssi_90['rssi_90'])))
print("packet loss probbility: "+str((1-rssi_90.count()['id_90']/data_num)*100)+"%")
plt.figure()
rssi_90['rssi_90'].plot()
plt.show()
rssi_90['rssi_90'].plot.box()
plt.show()
```

```python
before = 0
for i in rssi_90['id_90']:
    if before != i -1:
        print(before,i)
    before = i
```

```python
combined_rssi = pd.concat([rssi_0['rssi_0'],rssi_90['rssi_90']], axis=1)
combined_rssi.head()
combined_rssi.plot.box()
```

```python
data_num = 100
packetloss = pd.DataFrame(np.array([(1-combined_rssi.count()['rssi_0']/data_num)*100,(1-combined_rssi.count()['rssi_90']/data_num)*100]),
    index=pd.Index([np.mean(combined_rssi['rssi_0']),np.mean(combined_rssi['rssi_90'])]))
packetloss.plot(legend=None,xlim=[np.mean(combined_rssi['rssi_0'])+5,np.mean(combined_rssi['rssi_90'])-5])
packetloss
```
