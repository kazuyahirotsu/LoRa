```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get rssi value of 0001 and plot
value = pd.read_csv('0001_RSSI-data-2022-12-06 15_39_29.csv',  sep=',', usecols=[2, 3], names=('datetime', 'value'), skiprows=1)
value['datetime'] = pd.to_datetime(value['datetime'])
value = value.set_index('datetime').sort_index()
value["2022-11-17":"2022-12-05"].plot(style="o", ms=0.5,figsize=(20,2),ylim=(-150,0))

# get sunny time data
sunny_time = pd.read_csv('sunny_time_11_16_to_12_5.csv',  sep=',', usecols=[0, 1], names=('datetime', 'sunny_time'))
sunny_time = sunny_time.fillna(0)
sunny_time['datetime'] = pd.to_datetime(sunny_time['datetime'])

# merge 2 data and plot
merged_data = pd.merge(sunny_time, value, on='datetime', how='outer')
merged_data = merged_data.set_index('datetime').sort_index()
merged_data = merged_data.fillna({'sunny_time': 0})
axes = merged_data["2022-11-17":"2022-12-05"].plot(subplots=True, figsize=(20,4))
axes[1].set_ylim((-150,0))
plt.savefig("result0001.png", facecolor="white")
```
