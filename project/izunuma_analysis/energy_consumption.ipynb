```python
import pandas as pd
energy_end = pd.read_csv('energy_end2.csv',  sep=',', usecols=[0, 10], names=('time', 'power'), skiprows=1)
energy_end = energy_end.set_index('time').sort_index()
energy_end.head()
energy_end["power"][:"10:00.0"].plot(figsize=(20,4),ylabel="power (W)")
```

```python
import copy
energy_end_bad = copy.copy(energy_end)
energy_end_bad["power"]["00:21.8":] = 0.7247
energy_end_bad["power"][:"10:00.0"].plot(figsize=(20,4),ylabel="power (W)")
```
